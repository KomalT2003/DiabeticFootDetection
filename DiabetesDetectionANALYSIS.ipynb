```python
# For Health Issues add columns using classes
# add columns with prefix Health Issue - : Kidney Problem, Lung Problem , Muscle Weakness, Tingling Sensation, Thyroid, Paralysis, eyesight or vision, covid 19 infection, deep wound
# Blood Pressure, Heart Problem, Joint Pains, Cholestrol, Cancer, Liver Problems, Increased Body Weights

from sklearn.preprocessing import MultiLabelBinarizer

health_issues = [
    'Kidney Problem', 'Lung Problem', 'Muscle Weakness', 'Tingling Sensation', 'Thyroid', 'Paralysis',
    'Eyesight or Vision', 'COVID-19 Infection', 'Deep Wound', 'Blood Pressure', 'Heart Problem',
    'Joint Pains', 'Cholestrol', 'Cancer', 'Liver Problems', 'Increased Body Weights'
]

# Replace NaN with empty lists
df['Health Issues'] = df['Health Issues'].apply(lambda x: x if isinstance(x, list) else [])

# Initialize MultiLabelBinarizer with the specified health issues
mlb = MultiLabelBinarizer(classes=health_issues)

# Fit and transform the Health Issues column
encoded_health_issues = mlb.fit_transform(df['Health Issues'])

# Convert to DataFrame and add prefix
encoded_df = pd.DataFrame(encoded_health_issues, columns=[f'Health Issue - {issue}' for issue in mlb.classes_])

# Concatenate the original DataFrame with the encoded one
df = pd.concat([df, encoded_df], axis=1)
```

```python


from sklearn.preprocessing import MultiLabelBinarizer

health_conditions = [
    'blood pressure', 'blood sugar', 'cholestrol', 'heart problem', 'thyroid problem'
]

df['Health Condition'] = df['Health Condition'].apply(lambda x: x if isinstance(x, list) else [])

mlb = MultiLabelBinarizer(classes=health_conditions)

encoded_health_conditions = mlb.fit_transform(df['Health Condition'])

encoded_df = pd.DataFrame(encoded_health_conditions, columns=[f'Health Condition - {issue}' for issue in mlb.classes_])

df = pd.concat([df, encoded_df], axis=1)
```

```python
# Handling age column
# Encode date ranges 18-44 as 1, 45-54 as 2, 55-64 as 3, >=65 as 4
df['Age'] = df['Age'].replace(['18-44', '45-54', '55-64', '>=65'], [1, 2, 3, 4])
```

```python
# Handling Gender column
# Encode Male as 1 and Female as 0 other as 2
df['Gender'] = df['Gender'].replace(['Male', 'Female', 'other'], [1, 0, 2])
```

```python
# Handling Working Column
# Encode Student as 0, Working as 1, Retired as 2
df['Working'] = df['Working'].replace(['Student', 'Working', 'Retired'], [0, 1, 2])
df['Working'] = df['Working'].replace(['student', 'working', 'retired'], [0, 1, 2])

# Drop rows with NaN in Working column
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Working'])

# Convert Working column to integer
df['Working'] = df['Working'].astype(int)
```

```python
# Handling Height Column

# print all values in height column
print(df['Height'].unique())

# convert to int by rounding off
df['Height'] = df['Height'].astype(int)
```

```python
# Handling Weight Column
# remove kg and kgs from data
df['Weight'] = df['Weight'].str.replace('kg', '')
df['Weight'] = df['Weight'].str.replace('kgs', '')
df['Weight'] = df['Weight'].str.replace('Kg', '')
df['Weight'] = df['Weight'].str.replace('s', '')
df['Weight'] = df['Weight'].str.replace(' ', '')


# print all values in weight column
print(df['Weight'].unique())

# convert to float type
df['Weight'] = df['Weight'].astype(float)

# print all values in weight column
print(df['Weight'].unique())
```

```python
# Handling Diabetes column
# Encode Yes as 1 and No as 0
df['Diabetes'] = df['Diabetes'].replace(['Yes', 'No'], [1, 0])
```

```python
# Handling Blood Sugar column
# Encode Yes as 1 and No as 0
df['Blood Sugar'] = df['Blood Sugar'].replace(['Yes', 'No'], [1, 0])
```

```python
# Handling Close Relative column
# Encode Yes as 1 and No as 0
df['Close Relative'] = df['Close Relative'].replace(['Yes', 'No'], [1, 0])

# Handling Family History column
# Encode Yes as 1 and No as 0
df['Family History'] = df['Family History'].replace(['Yes', 'No'], [1, 0])
```

```python
# Handling Waist Circum Range
print(df['Waist Circum Range'].unique())

# encode less than 80 am ( <31 inches) as 1
# encode 80 - 88 cm ( 31-34 inches) as 2
# encode 89 - 93 cm (35-36 inches) as 3
# encode 94 - 102 cm (37- 40 inches) as 4
# encode greater than 102 cm (40 inches) as 5
df['Waist Circum Range'] = df['Waist Circum Range'].replace(['less than 80 am ( <31 inches)', '80 - 88 cm ( 35-36 inches)','89 - 93 cm (35-36 inches)', '94 - 102 cm (37- 40 inches)', 'greater than 102 cm (40 inches)'], [1, 2, 3, 4, 5])
```

```python
# Handling Health Issues
# print unique values
print(df['Health Issues'].unique())

# Create columns named all these unique values inside the array
df = pd.concat([df, df['Health Issues'].str.get_dummies(sep=', ')], axis=1)

```

```python
# Handle steroid column
# Encode Yes as 1 and No as 0
df['Steroid'] = df['Steroid'].replace(['Yes', 'No'], [1, 0])

# drop records with values null
df = df.dropna(subset=['Steroid'])

# convert steroid column to int
df['Steroid'] = df['Steroid'].astype(int)
```

```python
# Handle Low/Stressed column
# Encode never as 0, rarely as 1, often as 2, most of the time as 3, everyday as 4
df['Low/Stressed'] = df['Low/Stressed'].replace(['never', 'rarely', 'often', 'most of the time', 'everyday'], [0, 1, 2, 3, 4])
```

```python
# Handle Smoking column
# encode no as 0, yes as 2, was smoking earlier stopped for more than six months as 1
df['Smoking'] = df['Smoking'].replace(['no', 'yes', 'was smoking earlier stopped for more than six months'], [0, 2, 1])
```
